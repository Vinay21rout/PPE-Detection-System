# for splitting the data as 20% for validation & 10% for testing out of only trained data

```python
import os
import random
import shutil

# Source directories
image_dir = "datasets/images/train"
label_dir = "datasets/labels/train"

# Validation directories
val_image_dir = "datasets/images/val"
val_label_dir = "datasets/labels/val"

# Test directories
test_image_dir = "datasets/images/test"
test_label_dir = "datasets/labels/test"

# Get all images
images = os.listdir(image_dir)

# Shuffle randomly
random.shuffle(images)

# Dataset sizes
total_images = len(images)

val_split = int(0.2 * total_images)   # 20% validation
test_split = int(0.1 * total_images)  # 10% testing

# Split image lists
val_images = images[:val_split]
test_images = images[val_split:val_split + test_split]

# Function to move image + label
def move_files(image_list, target_image_dir, target_label_dir):

    for image in image_list:

        # Move image
        shutil.move(
            os.path.join(image_dir, image),
            os.path.join(target_image_dir, image)
        )

        # Matching label
        label = image.rsplit('.', 1)[0] + ".txt"

        if os.path.exists(os.path.join(label_dir, label)):
            shutil.move(
                os.path.join(label_dir, label),
                os.path.join(target_label_dir, label)
            )

# Move validation files
move_files(val_images, val_image_dir, val_label_dir)

# Move test files
move_files(test_images, test_image_dir, test_label_dir)

print("Train/Validation/Test split created successfully!")
```
